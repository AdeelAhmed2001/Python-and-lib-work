```python
# Multiple Points Graph With label and Title 
```

```python
x2points=np.array([2,6,9,10])
y2points=np.array([10,20,30,40])
```

```python
plt.xlabel("x axis horizontal")
plt.ylabel("y axis vertical")
plt.title("graph of matplotlib")

plt.plot(x2points,y2points)
plt.show()
```

```python
# Markers And Colors
```

```python

plt.plot(x2points,y2points,marker="o")
plt.show()
```

```python
plt.xlabel("x axis horizontal")
plt.ylabel("y axis vertical")
plt.title("graph of matplotlib")
plt.plot(x2points,y2points,marker="*",c="k")
plt.show()
```

```python
# If X axis is not provide so by default X axis is 0,1,2,3,4 etc  
```

```python
plt.plot(y2points,"*:c")
plt.show()
```

```python
# Marker size 
```

```python
plt.plot(x2points,y2points,marker="o",ms="20")
plt.show()
```

```python
# Linestyle
# "Short keys"
# Solid line (-)
# Dotted line (:)
# Dashed line (--) 
# Dashed/Dotted line (-.)
```

```python
plt.plot(x2points,y2points,marker="o",ms="20",ls="dashed")
plt.show()
```

```python
plt.plot(x2points,y2points,marker="o",ms="20",ls="-.")
plt.show()
```

```python
# Line Width
```

```python
plt.plot(x2points,y2points,marker="o",ms="20",lw="10")
plt.show()
```

```python
# Grid
```

```python
plt.grid()
```

```python
plt.plot(x2points,y2points,marker="o",ms="20",lw="10")
plt.grid()
```

```python
# If you want grid lines only one axis 
plt.plot(x2points,y2points,marker="o")
plt.grid(axis="y")
```

```python
# Displaying multiple plot with subplot function
# subplot function takes 3 arguments
# 1st arg=row , 2nd arg=column , 3rd arg=diagram number  
```

```python
# plot 1
x=np.array([0,1,2,3])
y=np.array([3,6,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)

# plot 2
x2=np.array([0,1,2,3])
y2=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.show()
```

```python
# creating scatter plot
```

```python
x=np.array([4,5,23,24,5,4,53,5,6,7])
y=np.array([43,43,64,98,76,85,62,11,23,66])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
# color and size change of scatter plot
```

```python
plt.scatter(x,y,c="y",s=60)
plt.show()
```

```python
sizes=np.array([30,60,80,90,100,150,200,250,400,130])
plt.scatter(x,y,c="y",s=sizes)
plt.show()
```

```python
# Creating bar plot
```

```python
a=np.array(["A","B","C","D"])
b=np.array([4,10,20,44])
plt.bar(a,b)
plt.show()
```

```python
# Horizontal bar
```

```python
a=np.array(["A","B","C","D"])
b=np.array([4,10,20,44])
plt.barh(a,b)
plt.show()
```

```python
# Creating pie charts with pie function
# it takes only one argument  "x"  or either  "y"
```

```python
y 
```

```python
plt.pie(y)
plt.show()
```

```python
# Labels and Shadow 
```

```python
mylabels=["Apple","Banana","Orange","Mango","cherry"]
plt.pie(y,labels=mylabels)
plt.show()
```

```python
mylabels=["Apple","Banana","Orange","Mango","cherry"]
plt.pie(y,labels=mylabels,shadow="True")
plt.show()
```

```python
mylabels=["Apple","Banana","Orange","Mango","cherry"]
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```
